# Train YOLOv8 Model
Backbone (Feature Extractor) â€“ Applies convolutions on the input image to capture low-level features like edges, textures, and basic shapes.

Neck (Feature Aggregator) â€“ Combines features from different layers/scales of the backbone to create multi-scale representations that help detect objects of various sizes.

Head (Prediction Layer) â€“ Uses convolution on the aggregated features to predict bounding boxes, class labels, and confidence scores.

```python
#Confusion Matrix Calculation
import numpy as np
import pandas as pd

#Your confusion matrix
cm = np.array(
  [ [32, 0, 0, 5],
 [0, 134, 0, 14],
 [0, 0, 50, 9],
 [0, 1, 0, 0]
]
)

num_classes = cm.shape[0] - 1  # exclude background
class_names = [f"Class_{i}" for i in range(num_classes)]

results = []

for i in range(num_classes):
    TP = cm[i, i]
    FN = cm[i, -1]                      # missed detections
    FP = cm[-1, i]                      # false alarms
    TN = cm.sum() - (TP + FP + FN)

    precision = TP / (TP + FP + 1e-6)
    recall = TP / (TP + FN + 1e-6)
    f1 = 2 * precision * recall / (precision + recall + 1e-6)

    results.append([
        class_names[i],
        TP, FP, FN, TN,
        precision, recall, f1
    ])

df = pd.DataFrame(
    results,
    columns=[
        "Class", "TP", "FP", "FN", "TN",
        "Precision", "Recall", "F1-score"
    ]
)

print(df)
```
